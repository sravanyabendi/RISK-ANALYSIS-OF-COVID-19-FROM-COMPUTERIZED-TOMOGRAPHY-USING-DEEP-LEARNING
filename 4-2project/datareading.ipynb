```python
path=r'C:\Users\sudhakararao\OneDrive\Desktop\Project\divideddataset'
l=['negative','generated_negative_images','positive-hometreat','positive-ICU','positive-hospitalized']
dic={'negative':0,'generated_negative_images':0,'positive-hometreat':1,'positive-ICU':1,'positive-hospitalized':1}
```

```python
alldata=[]
for i in l:
    c=0
    dataset=os.path.join(path,i)
    #print(dataset)
    for j in os.listdir(dataset):
        imagepath=os.path.join(dataset,j)
        #print(imagepath)
        img=cv2.imread(imagepath)
        #print(img,img.shape)
        img=cv2.resize(img,(128,128))
        plt.imshow(img)
        alldata.append([img,dic[i]])
        print(i,c)
        c+=1
```

```python
import random
random.shuffle(alldata)
```

```python
x=[]
y=[]
for a,b in alldata:
    x.append(a)
    y.append(b)
    print(b,end=" ")
```

```python
x=numpy.array(x)
y=numpy.array(y)
print(x.shape,y.shape)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
pickle.dump(xtrain,open('xtrain','wb'))
pickle.dump(ytrain,open('ytrain','wb'))
pickle.dump(xtest,open('xtest','wb'))
pickle.dump(ytest,open('ytest','wb'))
```
